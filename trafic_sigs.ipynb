```python
import pandas as pd
import os

# Pfade zu den Bild- und Label-Verzeichnissen
csv_path = 'C:/Uni/python/trafic sign/dataset/Train.csv'
base_image_folder = 'C:/Uni/python/trafic sign/dataset'  # Basispfad zu den Bildern (ohne "Train")
label_dir = 'C:/Uni/python/trafic sign/dataset/labels'  # Pfad zu den YOLO-Labels

# Stelle sicher, dass das Label-Verzeichnis existiert
os.makedirs(label_dir, exist_ok=True)

# Funktion zum Umwandeln der Bounding Box in YOLO-Format
def convert_to_yolo_format(row, img_width, img_height):
    x_center = (row['Roi.X1'] + row['Roi.X2']) / 2.0 / img_width
    y_center = (row['Roi.Y1'] + row['Roi.Y2']) / 2.0 / img_height
    width = (row['Roi.X2'] - row['Roi.X1']) / img_width
    height = (row['Roi.Y2'] - row['Roi.Y1']) / img_height
    return row['ClassId'], x_center, y_center, width, height

# Lade die CSV-Datei mit den Bounding Boxes
data = pd.read_csv(csv_path)

# Konvertiere jede Zeile der CSV in YOLO-Labels und speichere als .txt
for index, row in data.iterrows():
    # Erstelle den Pfad zum Bild (ohne doppeltes "Train")
    img_path = os.path.join(base_image_folder, row['Path']).replace('\\', '/')

    # Überprüfe, ob das Bild existiert
    if os.path.exists(img_path):
        # Hole die Bildgröße
        img_width, img_height = row['Width'], row['Height']
        
        # Konvertiere Bounding Box in YOLO-Format
        class_id, x_center, y_center, width, height = convert_to_yolo_format(row, img_width, img_height)
        
        # Pfad zur Speicherung der Label-Dateien
        label_subdir = os.path.join(label_dir, os.path.dirname(row['Path']))
        os.makedirs(label_subdir, exist_ok=True)
        
        # Erstelle den Pfad zur .txt Datei
        label_file_path = os.path.join(label_subdir, os.path.basename(row['Path']).replace('.png', '.txt'))
        with open(label_file_path, 'w') as f:
            f.write(f"{class_id} {x_center:.6f} {y_center:.6f} {width:.6f} {height:.6f}\n")

        print(f"Label gespeichert: {label_file_path}")
    else:
        print(f"Bild nicht gefunden: {img_path}")
```
